```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datset = pd.read_csv("twitter_training.csv")
print(datset.shape)
datset.head(10)
```

Adding column names

```python
datset.columns = ['ID', 'Brand', 'Sentiment', 'Comments']
datset.columns
datset.head(10)
# datset.describe
```

Checking null values

```python
datset.isnull().sum()
```

```python
datset.dropna(inplace=True)
datset.duplicated().sum()
```

```python
datset.drop_duplicates(inplace = True)
datset.duplicated().sum()
```

```python
datset.info()
datset.describe(include=object)
```

```python
datset['Brand'].value_counts()
```

```python
datset['Sentiment'].unique()
```

```python
ar = datset['Brand'].unique()
print("Total Brands", len(ar))
```

```python
plt.figure(figsize=(14,14))
ax = sns.countplot(y='Brand', data= datset, order= datset['Brand'].value_counts(ascending= False).index, palette= 'Paired')
for i in ax.containers:
    ax.bar_label(i)
plt.title("Brands Count")
plt.show() 
```

```python
ax = sns.countplot(x= 'Sentiment', data= datset, palette= 'deep')
for i in ax.containers:
    ax.bar_label(i)
plt.xlabel('Sentiments')
plt.ylabel('N.o of users')
plt.show()
```

```python
sen_count = datset['Sentiment'].value_counts()
plt.pie(sen_count, labels= sen_count.index, autopct= "%1.1f%%")
plt.title("Overall Sentiment Distribution")
plt.show()
```

```python
plt.figure(figsize= (14,14))
sns.countplot(x = 'Brand', data= datset, hue= "Sentiment", palette= "viridis")
plt.xticks(rotation = 90)
plt.title("Braand-wise Sentiment Distribution")
plt.show()
```

```python
brand_wise_sentiment = datset.groupby(["Brand", "Sentiment"]).size().reset_index(name='Count')
brand_count = datset["Brand"].value_counts().nlargest(7).index
# brand_count
top_brand_sentiment = brand_wise_sentiment[brand_wise_sentiment['Brand'].isin(brand_count)]
plt.figure(figsize=(10,5))
ax = sns.barplot(data=top_brand_sentiment[top_brand_sentiment['Sentiment'] == 'Negative'], x='Brand', y='Count', palette='viridis')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Top 7 Brands with Negative Sentiments')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=top_brand_sentiment[top_brand_sentiment['Sentiment'] == 'Positive'], x='Brand', y='Count', palette='viridis')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Top 7 Brands with Positive Sentiments')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=top_brand_sentiment[top_brand_sentiment['Sentiment'] == 'Neutral'], x='Brand', y='Count', palette='Greys')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Top 7 Brands with Neutral Sentiments')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(data=top_brand_sentiment[top_brand_sentiment['Sentiment'] == 'Irrelevant'], x='Brand', y='Count', palette='Blues')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Top 7 Brands with Irrelevant Sentiments')
plt.xlabel('Brands')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Sentiment Distribution for world of craft

```python
plt.figure(figsize=(5,5))
wld_of_cft = datset[datset["Brand"] == "WorldOfCraft"]
senti_count = wld_of_cft["Sentiment"].value_counts()
print(senti_count)

plt.pie(senti_count, labels=senti_count.index, autopct="%1.1f%%")
plt.title("Sentiment distribution for WorldOfCraft")
plt.show()
```

```python
plt.figure(figsize=(5,5))

nvidia = datset[datset["Brand"] == "Nvidia"]
senti_count = nvidia["Sentiment"].value_counts()
print(senti_count)


plt.pie(senti_count, labels=senti_count.index, autopct="%1.1f%%", startangle=160)
plt.title("Sentiment distribution for Nvidia")
plt.show()
```

```python
plt.figure(figsize=(5,5))

microsoft = datset[datset["Brand"] == "Microsoft"]
senti_count = microsoft["Sentiment"].value_counts()
print(senti_count)

plt.pie(senti_count, labels=senti_count.index, autopct="%1.1f%%", startangle=160)
plt.title("Sentiment distribution for Microsoft")
plt.show()
```

```python
datset["Msg_len"] = datset["Comments"].apply(len)
datset
```

```python
sns.boxplot(data=datset, x=datset['Sentiment'], y='Msg_len', palette='deep', order=['Positive', 'Negative', 'Neutral', 'Irrelevant'])
plt.title('Message Length Distribution by Sentiment')
plt.ylabel('Message Length')
plt.xlabel('Sentiment')
plt.ylim(0,250)
plt.show()  
```
